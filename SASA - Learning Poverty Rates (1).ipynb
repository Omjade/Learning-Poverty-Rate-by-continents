# North America - Learning Poverty Percentage 


```python
import pandas as pd

data = {
    "Country": [
        "Belize", "Canada", "Costa Rica", "Cuba", "Dominican Republic", 
        "El Salvador", "Grenada", "Guatemala", "Haiti", "Honduras", 
        "Jamaica", "Mexico", "Nicaragua", "Panama", 
        "Trinidad and Tobago", "United States of America (USA)"
    ],
    "Learning Poverty (%)": [
        76.4, 4.3, 34, 24.8, 77.7, 69.1, 3.2, 78.5, 42.7, 
        79.3, 7.7, 43.2, 69.8, 78.5, 20.6, 4.3
    ]
}

northamerica_learning_poverty = pd.DataFrame(data)


northamerica_learning_poverty.head(10)
```

#
Learning Poverty % = % of children aged 10 who cannot read and understand a simple age-appropriate text.

| **Learning Poverty %** | **What It Means**                                                        |
| ---------------------- | ------------------------------------------------------------------------ |
| **80%+**               | Most children (8 in 10) can’t read properly — **severe learning crisis** |
| **50–79%**             | More than half are struggling — **high concern**                         |
| **20–49%**             | Moderate concern — **room for improvement**                              |
| **<20%**               | Most kids can read — **strong education system**                         |
| **<5%**                | Very low — **excellent literacy outcomes**                               |



```python
import matplotlib.pyplot as plt

# Sort the DataFrame
sorted_df = northamerica_learning_poverty.sort_values(by="Learning Poverty (%)", ascending=False)

# Calculate the average
average = sorted_df["Learning Poverty (%)"].mean()

# Plot setup
plt.figure(figsize=(14, 7))
bars = plt.bar(sorted_df["Country"], sorted_df["Learning Poverty (%)"], color="#69b3a2", edgecolor="black")

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.1f}%',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),
                 textcoords="offset points",
                 ha='center', va='bottom',
                 fontsize=10, fontweight='bold')

# Average line
plt.axhline(average, color='red', linestyle='--', linewidth=1.5, label=f'Average: {average:.1f}%')
plt.legend()

# Titles and labels
plt.title("📉 Learning Poverty Rates in North American Countries", fontsize=16, fontweight='bold')
plt.ylabel("Learning Poverty (%)", fontsize=12)
plt.xlabel("Country", fontsize=12)
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
top_8 = sorted_df.head(8)
plt.figure(figsize=(8, 8))
plt.pie(top_8["Learning Poverty (%)"], labels=top_8["Country"], autopct='%1.1f%%', startangle=140)
plt.title("Top 8 Countries by Learning Poverty (%)", fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
oceania_data = {
    "Country": ["Australia", "New Zealand", "Nauru", "Papua New Guinea"],
    "Learning Poverty (%)": [9, 11, 70, 94]
}

# DataFrame
oceania_df = pd.DataFrame(oceania_data)

# Ensure numeric values
oceania_df["Learning Poverty (%)"] = pd.to_numeric(oceania_df["Learning Poverty (%)"], errors='coerce')

# Sort
sorted_oceania_df = oceania_df.sort_values(by="Learning Poverty (%)", ascending=False)

# Line Chart
plt.figure(figsize=(8, 5))
plt.plot(sorted_oceania_df["Country"].values,  # Convert to numpy array
         sorted_oceania_df["Learning Poverty (%)"].values,
         marker='o', linestyle='-', color='purple')
plt.title("Learning Poverty Trend in Australia & Oceania", fontsize=14)
plt.xlabel("Country")
plt.ylabel("Learning Poverty (%)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
