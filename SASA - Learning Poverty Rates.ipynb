# North America - Learning Poverty Percentage 


```python
import pandas as pd

data = {
    "Country": [
        "Belize", "Canada", "Costa Rica", "Cuba", "Dominican Republic", 
        "El Salvador", "Grenada", "Guatemala", "Haiti", "Honduras", 
        "Jamaica", "Mexico", "Nicaragua", "Panama", 
        "Trinidad and Tobago", "United States of America (USA)"
    ],
    "Learning Poverty (%)": [
        76.4, 4.3, 34, 24.8, 77.7, 69.1, 3.2, 78.5, 42.7, 
        79.3, 7.7, 43.2, 69.8, 78.5, 20.6, 4.3
    ]
}

northamerica_learning_poverty = pd.DataFrame(data)


northamerica_learning_poverty.head(10)
```

#
Learning Poverty % = % of children aged 10 who cannot read and understand a simple age-appropriate text.

| **Learning Poverty %** | **What It Means**                                                        |
| ---------------------- | ------------------------------------------------------------------------ |
| **80%+**               | Most children (8 in 10) can’t read properly — **severe learning crisis** |
| **50–79%**             | More than half are struggling — **high concern**                         |
| **20–49%**             | Moderate concern — **room for improvement**                              |
| **<20%**               | Most kids can read — **strong education system**                         |
| **<5%**                | Very low — **excellent literacy outcomes**                               |



```python
import matplotlib.pyplot as plt

# Sort the DataFrame
sorted_df = northamerica_learning_poverty.sort_values(by="Learning Poverty (%)", ascending=False)

# Calculate the average
average = sorted_df["Learning Poverty (%)"].mean()

# Plot setup
plt.figure(figsize=(14, 7))
bars = plt.bar(sorted_df["Country"], sorted_df["Learning Poverty (%)"], color="#69b3a2", edgecolor="black")

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.1f}%',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),
                 textcoords="offset points",
                 ha='center', va='bottom',
                 fontsize=10, fontweight='bold')

# Average line
plt.axhline(average, color='red', linestyle='--', linewidth=1.5, label=f'Average: {average:.1f}%')
plt.legend()

# Titles and labels
plt.title("📉 Learning Poverty Rates in North American Countries", fontsize=16, fontweight='bold')
plt.ylabel("Learning Poverty (%)", fontsize=12)
plt.xlabel("Country", fontsize=12)
plt.xticks(rotation=45, ha="right", fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
top_8 = sorted_df.head(8)
plt.figure(figsize=(8, 8))
plt.pie(top_8["Learning Poverty (%)"], labels=top_8["Country"], autopct='%1.1f%%', startangle=140)
plt.title("Top 8 Countries by Learning Poverty (%)", fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Raw data as dictionary
europe_data = {
    "Country": [
        "Armenia", "Austria", "Azerbaijan", "Belgium", "Bosnia and Herzegovina", "Bulgaria", "Croatia",
        "Cyprus", "Czechia", "Denmark", "Finland", "France", "Georgia", "Germany", "Greece", "Hungary",
        "Iceland", "Ireland", "Italy", "Kazakhstan", "Latvia", "Lithuania", "Luxembourg", "Malta",
        "Moldova", "Montenegro", "Netherlands", "North Macedonia", "Norway", "Poland", "Portugal",
        "Romania", "Russia", "Serbia", "Slovakia", "Slovenia", "Spain", "Sweden", "Turkey", "Ukraine",
        "United Kingdom (UK)"
    ],
    "Learning Poverty (%)": [
        27.2, 15, 37.5, 6, 28, 13, 4,
        16, 3, 4, 3, 7, 15, 6, 8, 9.2,
        9, 2, 3, 2, 4, 3.3, 3, 28,
        11, 28, 2, 40, 6, 5, 6,
        16, 3.8, 8, 9, 6, 5, 2, 15, 28,
        4
    ]
}

# Create DataFrame
europe_df = pd.DataFrame(europe_data)

# Sort by Learning Poverty
sorted_europe_df = europe_df.sort_values(by="Learning Poverty (%)", ascending=False)

# -------- BAR CHART WITH LABELS --------
plt.figure(figsize=(14, 7))
bars = plt.bar(sorted_europe_df["Country"], sorted_europe_df["Learning Poverty (%)"], edgecolor='black', color='skyblue')

plt.title("Learning Poverty (%) in European Countries", fontsize=16)
plt.xlabel("Country")
plt.ylabel("Learning Poverty (%)")
plt.xticks(rotation=75, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{height:.1f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Convert to NumPy arrays to avoid multi-dimensional indexing error
x_values = np.array(sorted_europe_df["Country"].tolist())
y_values = np.array(sorted_europe_df["Learning Poverty (%)"].tolist())

# Plot the line chart
plt.figure(figsize=(14, 6))
plt.plot(x_values, y_values, marker='o', linestyle='-', color='purple')
plt.title("Trend of Learning Poverty in Europe", fontsize=16)
plt.xlabel("Country")
plt.ylabel("Learning Poverty (%)")
plt.xticks(rotation=75, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
